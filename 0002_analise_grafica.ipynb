```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st

%matplotlib inline
sns.set(color_codes = True)
```

```python
CS = pd.read_excel("https://github.com/julianovale/lean_6_sigma_python/blob/master/descstats.xlsx?raw=true")
```

```python
CS.head()
```

```python
CS = CS.dropna(axis = 1)
CS
```

# Criando Boxplot

```python
sns.boxplot(CS['Sample 1'])
```

```python
sns.boxplot(CS['Sample 1'], orient = "v")
```

## exportando uma imagem do boxplot

```python
bplot = sns.boxplot(CS['Sample 1'], orient = "v")
```

```python
plot_file_name = "Boxplot.jpg"
bplot.figure.savefig(plot_file_name, format = 'jpeg', dpi = 100)
```

```python
sns.boxplot(CS['Sample 1'], orient = "v", width = 0.3)
```

```python
b = CS[['Sample 1', 'Sample 2']]
```

```python
b.head()
```

```python
sns.boxplot(x='variable', y = 'value', data = pd.melt(b), orient = "v", width = 0.3)
```

```python
sns.boxplot(x=CS['Group'], y=CS['Sample 1'], orient = "v", width = 0.2)
```

```python
sns.boxplot(x=CS['Group'], y=CS['Sample 2'], orient = "v", width = 0.2)
```
